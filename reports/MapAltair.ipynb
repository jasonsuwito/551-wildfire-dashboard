```python
import pandas as pd
import altair as alt
import plotly.express as px
import dash
from dash import dcc, html
```

```python
fires = pd.read_csv("NFDB_point_20240613.txt")

fires.info()
```

```python
fire_map_data = fires[['FIRE_ID', 'SRC_AGENCY','LATITUDE', 'LONGITUDE', 'YEAR', 'SIZE_HA']]
fire_map_data = fire_map_data[(fire_map_data['YEAR'] >= 2013) & (fire_map_data['YEAR'] <= 2023)].sort_values(by='YEAR', ascending=True)

fire_map_data.head()
```

```python
fire_map_data.tail()
```

```python
import plotly.express as px
# First map (fire data layer)
fig = px.scatter_map(
    fire_map_data,
    lat="LATITUDE",
    lon="LONGITUDE",
    size="SIZE_HA",
    zoom=3.25,  
    center={"lat": 60, "lon": -100},
    size_max=25
)

# Update the first map trace (fire data)
fig.update_traces(
    marker=dict(color="orange", opacity=0.4),
    hoverinfo="skip",
    hovertemplate=None  
)

# Update layout
fig.update_layout(
    map_style="dark",
    autosize=True,
    width=1300, height=700,  
    margin={"r":0,"t":0,"l":0,"b":0},
    dragmode=False,
)

# Show the map
fig.show()
```

```python
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div([
    dcc.Graph(figure=fig)  # Embed the Plotly figure inside a Dash Graph component
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
